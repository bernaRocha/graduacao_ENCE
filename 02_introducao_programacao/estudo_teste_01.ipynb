```python
f1 = m1 = f2 = m2 = f3 = m3 = 0

for i in range(151):
    sexo = input('Registre o sexo, M para masculino e F para Feminino ').upper()
    freq_esportes = int(input('Registre o quanto pratica esportes, 1 para nuncam 2 para s vez e 3 para diariamente '))
    
    if sexo == 'F':
        if freq_esportes == 1:
            f1 += 1
        elif freq_esportes == 2:
            f2 += 1
        elif freq_esportes == 3:
            f3 += 1

    elif sexo == 'M':
        if freq_esportes == 1:
            m1 += 1
        elif freq_esportes == 2:
            m2 += 1
        elif freq_esportes == 3:
            m3 += 1
        
print(f'''
{f1} {10*'-'} {m1}
{f2} {10*'-'} {m2}
{f3} {10*'-'} {m3}
''')        
```
