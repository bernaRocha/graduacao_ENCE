```python
matriz = [[1000000000, 1, 2, 5], [1,0,3,6]]

n_linhas = len(matriz)
n_colunas = len(matriz[0])

for i in range(n_linhas):
    for j in range(n_colunas):
        if matriz[i][j] >= 1000000:
            print('Essa matriz tem um elemento acima de 1 milhão')
```

```python
def imprimir_matriz_bonita(matriz):
    for i in range(len(matriz)):
        for j in range(len(matriz[0])):
            print(f"{matriz[i][j]:>6}", end="")
        print()

ordem = int(input('Escolha a ordem que deseja na matriz quadrada [até 50]: '))

while ordem >= 50:
    print('Tem que ser menor que 50, faça novamente')
    ordem = int(input('Escolha a ordem que deseja na matriz quadrada [até 50]: '))
     
mat = []

for i in range(ordem): 
    mat.append([0] * ordem)

print('\nMatriz com zeros:')

imprimir_matriz_bonita(mat)

## Colocando números na matriz
print('\nColocando números na matriz \n')
for i in range(ordem):
    for j in range(ordem):
        print(f'Elemento[{i},{j}] = ', end="")
        mat[i][j] = int(input())
        
print('\nMatriz digitada ')        
imprimir_matriz_bonita(mat)

## Reversa da matriz
reversa = [[mat[j][i] for j in range(len(mat))] for i in range(len(mat[0]))]

## Imprimindo a reversa bonitinho
print('\nImprimindo a reversa')

imprimir_matriz_bonita(reversa)

print('\nAo verificar a simetria das matrizes constata-se que: ')
if mat == reversa:
    print('\nSão simétricas!')
else:
    print('\nNÃO são simétricas!')           
```
