```python
def imprimir_matriz_bonita(matriz):
    for i in range(len(matriz)):
        for j in range(len(matriz[0])):
            print(f"{matriz[i][j]:>6}", end="")
        print()

ordem = int(input('Escolha a ordem que deseja na matriz quadrada [at√© 50]: '))

while ordem >= 50:
    print('Tem que ser menor que 50, fa√ßa novamente')
    ordem = int(input('Escolha a ordem que deseja na matriz quadrada [at√© 50]: '))
     
mat = []

for i in range(ordem): 
    mat.append([0] * ordem)

print('\nMatriz com zeros:')

imprimir_matriz_bonita(mat)

## Colocando n√∫meros na matriz
print('\nColocando n√∫meros na matriz \n')
for i in range(ordem):
    for j in range(ordem):
        print(f'Elemento[{i},{j}] = ', end="")
        mat[i][j] = int(input())
        
print('\nMatriz digitada ')        
imprimir_matriz_bonita(mat)

## Reversa da matriz
reversa = [[mat[j][i] for j in range(len(mat))] for i in range(len(mat[0]))]

## Imprimindo a reversa bonitinho
print('\nImprimindo a reversa')

imprimir_matriz_bonita(reversa)

print('\nAo verificar a simetria das matrizes constata-se que: ')
if mat == reversa:
    print('\nS√£o sim√©tricas')
else:
    print('\nN√ÉO s√£o sim√©tricas')           
```
