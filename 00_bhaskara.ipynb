# Fórmula de Bhaskara

```python
def bhaskara(a, b, c):
    """
    a, b e c são os coeficientes da equação do segundo grau a ser calculado as raízes
    """
    import math
    delta = b ** 2 - (4 * a * c)
    x1 = (- b + math.sqrt(delta))/ 2 * a
    x2 = (- b - math.sqrt(delta))/ 2 * a
    return x1, x2
```

```python
bhaskara(1, -4, 3)
```

```python
bhaskara(1, 2, -3)
```
