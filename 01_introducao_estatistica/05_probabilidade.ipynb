```python
from math import comb

tentativas = 5
sucessos = 3
fracasso = tentativas - sucessos
prob_1_sucesso = 0.5

probabilidade = comb(tentativas, sucessos) * (prob_1_sucesso ** sucessos) * ((1 - prob_1_sucesso) ** (fracasso))
probabilidade
```
